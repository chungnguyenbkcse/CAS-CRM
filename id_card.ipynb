```python
import cv2
from PIL import Image
from ultralytics import YOLO
import numpy as np

def safe_decode(qr_data_bytes):
    """Thá»­ decode dá»¯ liá»‡u QR báº±ng nhiá»u báº£ng mÃ£ Ä‘á»ƒ trÃ¡nh lá»—i font."""
    encodings = ['utf-8', 'windows-1258', 'latin1', 'big5', 'shift-jis']
    for enc in encodings:
        try:
            return qr_data_bytes.decode(enc)
        except:
            continue
    return qr_data_bytes.decode('utf-8', errors='ignore')

def detect_and_parse(image_path, model_path):
    # Load model YOLO
    model = YOLO(model_path)

    # Äá»c áº£nh
    img = Image.open(image_path)

    # Detect objects
    results = model.predict(source=image_path, conf=0.5, verbose=False)
    print(results[0].verbose())

    # DÃ¹ng OpenCV Ä‘á»c QR
    img_cv = cv2.imread(image_path)
    detector = cv2.QRCodeDetector()

    data, points, _ = detector.detectAndDecode(img_cv)
    if not data:
        print("âš ï¸ KhÃ´ng phÃ¡t hiá»‡n mÃ£ QR nÃ o.")
        return

    # Náº¿u lá»—i font, thá»­ decode láº¡i
    try:
        qr_raw = safe_decode(data.encode('latin1'))
    except Exception as e:
        qr_raw = data
        print("âš ï¸ KhÃ´ng thá»ƒ decode láº¡i:", e)

    print("âœ… QR raw:", qr_raw)

    # Parse theo Ä‘á»‹nh dáº¡ng CCCD Viá»‡t Nam (thÆ°á»ng 7 trÆ°á»ng, cÃ¡ch nhau báº±ng '|')
    parts = qr_raw.split('|')
    if len(parts) == 7:
        info = {
            'id': parts[0],
            'so': parts[1],
            'ho_ten': parts[2],
            'ngay_sinh': parts[3],
            'gioi_tinh': parts[4],
            'dia_chi': parts[5],
            'ngay_cap': parts[6]
        }
        print("ğŸ“„ ThÃ´ng tin CCCD:", info)
    else:
        print("âš ï¸ QR khÃ´ng Ä‘Ãºng Ä‘á»‹nh dáº¡ng CCCD Viá»‡t Nam!")
        print("ğŸ“„ ThÃ´ng tin CCCD:", None)

# === Gá»ŒI HÃ€M ===
detect_and_parse(
    image_path="/content/ID-Card-VietNam-2/test/images/cccd_me.jpg",
    model_path="/content/runs/detect/train/weights/best.pt"
)
```

```python
import cv2
import os
from ultralytics import YOLO
from qreader import QReader
from datetime import datetime
from PIL import Image

# =========================================================
# ğŸš€ STEP 3: Load model YOLO Ä‘Ã£ train
# =========================================================
model_path = "/content/runs/detect/train/weights/best.pt"  # ÄÆ°á»ng dáº«n tá»›i model Ä‘Ã£ train
assert os.path.exists(model_path), "âš ï¸ KhÃ´ng tÃ¬m tháº¥y file model best.pt!"
model = YOLO(model_path)

# =========================================================
# ğŸ§  HÃ€M 1: Cáº¯t vÃ¹ng QR tá»« áº£nh theo toáº¡ Ä‘á»™ YOLO
# =========================================================
def extract_qr_from_image(img, box):
    x, y, w, h = box
    crop = img[int(y):int(y+h), int(x):int(x+w)]
    qreader = QReader()
    decoded_text = qreader.detect_and_decode(image=crop)
    return decoded_text

# =========================================================
# ğŸ§  HÃ€M 2: Parse thÃ´ng tin CCCD tá»« QR code Viá»‡t Nam
# =========================================================
def parse_vietnam_id_qr(qr_text):
    """
    Chuáº©n QR CCCD gá»“m 7 pháº§n ngÄƒn cÃ¡ch bá»Ÿi dáº¥u |
    VÃ­ dá»¥: 075183018264|271524678|Nguyá»…n Thá»‹ Ngá»c Thuáº­n|06031983|Ná»¯|Tá»• 1, áº¤p Lá»™ 25, HÆ°ng Lá»™c, Thá»‘ng Nháº¥t, Äá»“ng Nai|12082021
    """
    if not qr_text:
        return None

    if isinstance(qr_text, tuple):
        qr_text = qr_text[0]  # bÃ³c ra pháº§n tá»­ Ä‘áº§u tiÃªn

    parts = qr_text.split("|")
    if len(parts) < 7:
        print("âš ï¸ QR khÃ´ng Ä‘Ãºng Ä‘á»‹nh dáº¡ng CCCD Viá»‡t Nam!")
        return None

    id_number = parts[0].strip()
    so = parts[1].strip()
    ho_ten = parts[2].strip()
    ngay_sinh = parts[3].strip()
    gioi_tinh = parts[4].strip()
    dia_chi = parts[5].strip()
    ngay_cap = parts[6].strip()

    # Äá»‹nh dáº¡ng láº¡i ngÃ y
    def fmt_date(s):
        try:
            return datetime.strptime(s, "%d%m%Y").strftime("%d/%m/%Y")
        except:
            return s

    return {
        "id": id_number,
        "so": so,
        "ho_ten": ho_ten,
        "ngay_sinh": fmt_date(ngay_sinh),
        "gioi_tinh": gioi_tinh,
        "dia_chi": dia_chi,
        "ngay_cap": fmt_date(ngay_cap),
    }

# =========================================================
# ğŸš€ STEP 4: DÃ² tÃ¬m QR vÃ  trÃ­ch xuáº¥t thÃ´ng tin
# =========================================================
image_path = "/content/ID-Card-VietNam-2/test/images/FILEUPLOAD_b94903b3-953d-4fd2-bf0d-181d746ea6b9_202507281045318210.jpg"
assert os.path.exists(image_path), "âš ï¸ KhÃ´ng tÃ¬m tháº¥y file áº£nh test!"
img = cv2.imread(image_path)

# Dá»± Ä‘oÃ¡n
results = model.predict(source=image_path, conf=0.4, iou=0.3)

# =========================================================
# ğŸš€ STEP 5: Láº·p qua cÃ¡c object Ä‘á»ƒ tÃ¬m QR_CODE
# =========================================================
found_qr = False
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])
        label = r.names[cls_id]

        if label.lower() == "qr_code":
            found_qr = True
            xyxy = box.xyxy[0].tolist()
            x1, y1, x2, y2 = xyxy
            w, h = x2 - x1, y2 - y1

            # Cáº¯t vÃ¹ng QR
            qr_text = extract_qr_from_image(img, (x1, y1, w, h))

            if isinstance(qr_text, tuple):
                qr_text = qr_text[0]

            if qr_text:
                print("âœ… QR raw:", qr_text)
                info = parse_vietnam_id_qr(qr_text)
                print("ğŸ“„ ThÃ´ng tin CCCD:", info)
            else:
                print("âš ï¸ KhÃ´ng Ä‘á»c Ä‘Æ°á»£c QR code, thá»­ áº£nh khÃ¡c hoáº·c rÃµ hÆ¡n!")

if not found_qr:
    print("âš ï¸ KhÃ´ng phÃ¡t hiá»‡n Ä‘Æ°á»£c vÃ¹ng QR trÃªn áº£nh!")
```

```python
import cv2
import os
from ultralytics import YOLO
from pyzbar.pyzbar import decode as pyzbar_decode
from datetime import datetime
from PIL import Image
import numpy as np

# =========================================================
# ğŸš€ STEP 3: Load model YOLO Ä‘Ã£ train
# =========================================================
model_path = "/content/runs/detect/train/weights/best.pt"  # ÄÆ°á»ng dáº«n tá»›i model Ä‘Ã£ train
assert os.path.exists(model_path), "âš ï¸ KhÃ´ng tÃ¬m tháº¥y file model best.pt!"
model = YOLO(model_path)

# =========================================================
# ğŸ§  HÃ€M 1: Cáº¯t vÃ¹ng QR tá»« áº£nh theo toáº¡ Ä‘á»™ YOLO
# =========================================================
def extract_qr_from_image(img, box):
    x, y, w, h = box
    crop = img[int(y):int(y+h), int(x):int(x+w)]

    # Má»Ÿ rá»™ng vÃ¹ng crop má»™t chÃºt Ä‘á»ƒ trÃ¡nh máº¥t viá»n QR
    padding = 10
    y1 = max(0, int(y) - padding)
    y2 = min(img.shape[0], int(y + h) + padding)
    x1 = max(0, int(x) - padding)
    x2 = min(img.shape[1], int(x + w) + padding)
    crop = img[y1:y2, x1:x2]

    # Upscale áº£nh QR lÃªn 3x Ä‘á»ƒ dá»… Ä‘á»c hÆ¡n
    scale_factor = 3
    h_new, w_new = crop.shape[0] * scale_factor, crop.shape[1] * scale_factor
    crop_large = cv2.resize(crop, (w_new, h_new), interpolation=cv2.INTER_CUBIC)

    # Thá»­ nhiá»u preprocessing methods
    def try_decode(image):
        # Method 1: DÃ¹ng pyzbar trá»±c tiáº¿p
        decoded_objects = pyzbar_decode(image)
        if decoded_objects:
            return decoded_objects[0].data

        # Method 2: DÃ¹ng OpenCV QRCodeDetector
        detector = cv2.QRCodeDetector()
        data, bbox, _ = detector.detectAndDecode(image)
        if data:
            return data.encode('utf-8')

        return None

    # 1. Thá»­ vá»›i áº£nh gá»‘c Ä‘Ã£ upscale
    result = try_decode(crop_large)
    if result:
        return decode_qr_data(result)

    # 2. Thá»­ vá»›i grayscale
    gray = cv2.cvtColor(crop_large, cv2.COLOR_BGR2GRAY)
    result = try_decode(gray)
    if result:
        return decode_qr_data(result)

    # 3. Thá»­ vá»›i adaptive threshold
    thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                     cv2.THRESH_BINARY, 11, 2)
    result = try_decode(thresh)
    if result:
        return decode_qr_data(result)

    # 4. Thá»­ vá»›i Otsu threshold
    _, otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    result = try_decode(otsu)
    if result:
        return decode_qr_data(result)

    # 5. Thá»­ lÃ m sáº¯c nÃ©t áº£nh
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    sharpened = cv2.filter2D(gray, -1, kernel)
    result = try_decode(sharpened)
    if result:
        return decode_qr_data(result)

    return None

def decode_qr_data(qr_data):
    """Decode QR data bytes sang text vá»›i encoding phÃ¹ há»£p"""
    if isinstance(qr_data, str):
        return qr_data

    # Debug: In ra bytes raw Ä‘á»ƒ xem
    print(f"ğŸ” DEBUG - Raw bytes (first 120): {qr_data[:120]}")
    print(f"ğŸ” DEBUG - Hex: {qr_data[:120].hex()}")

    # Chiáº¿n lÆ°á»£c: Decode UTF-8 vá»›i replace Ä‘á»ƒ giá»¯ cáº¥u trÃºc
    # Thay kÃ½ tá»± lá»—i báº±ng ï¿½ rá»“i xá»­ lÃ½ sau
    decoded = qr_data.decode('utf-8', errors='replace')
    print(f"ğŸ” UTF-8 vá»›i replace: {decoded}")

    # Náº¿u cÃ³ kÃ½ tá»± replacement ï¿½, thá»­ fix báº±ng cÃ¡ch decode láº¡i nhá»¯ng pháº§n cÃ³ thá»ƒ
    if 'ï¿½' in decoded:
        # TÃ¡ch theo dáº¥u | trÆ°á»›c
        parts = decoded.split('|')
        fixed_parts = []

        for i, part in enumerate(parts):
            if 'ï¿½' in part:
                # Thá»­ decode pháº§n nÃ y vá»›i cÃ¡c encoding khÃ¡c
                # TÃ¬m vá»‹ trÃ­ trong bytes gá»‘c
                start_idx = qr_data.find(part.encode('utf-8', errors='ignore'))

                # Vá»›i pháº§n giá»›i tÃ­nh thÆ°á»ng lÃ  "Nam" hoáº·c "Ná»¯"
                if i == 4 and part.startswith('N'):
                    # Thá»­ thay tháº¿
                    if part.strip().startswith('N'):
                        fixed_parts.append('Ná»¯')  # Giáº£ Ä‘á»‹nh lÃ  Ná»¯
                        print(f"ğŸ”§ Fixed part {i}: 'N?' -> 'Ná»¯'")
                        continue

                # Giá»¯ nguyÃªn pháº§n cÃ³ ï¿½, sáº½ fix sau
                fixed_parts.append(part)
            else:
                fixed_parts.append(part)

        result = '|'.join(fixed_parts)
        print(f"âœ… Decoded result: {result}")
        return result

    return decoded

# =========================================================
# ğŸ§  HÃ€M 2: Parse thÃ´ng tin CCCD tá»« QR code Viá»‡t Nam
# =========================================================
def parse_vietnam_id_qr(qr_text):
    """
    Chuáº©n QR CCCD gá»“m 7 pháº§n ngÄƒn cÃ¡ch bá»Ÿi dáº¥u |
    VÃ­ dá»¥: 075183018264|271524678|Nguyá»…n Thá»‹ Ngá»c Thuáº­n|06031983|Ná»¯|Tá»• 1, áº¤p Lá»™ 25, HÆ°ng Lá»™c, Thá»‘ng Nháº¥t, Äá»“ng Nai|12082021
    """
    if not qr_text:
        return None

    if isinstance(qr_text, tuple):
        qr_text = qr_text[0]  # bÃ³c ra pháº§n tá»­ Ä‘áº§u tiÃªn

    parts = qr_text.split("|")
    print(f"ğŸ” Sá»‘ pháº§n tÃ¡ch Ä‘Æ°á»£c: {len(parts)}")
    print(f"ğŸ” CÃ¡c pháº§n: {parts}")

    # Linh hoáº¡t: cháº¥p nháº­n 6-7 pháº§n (cÃ³ thá»ƒ thiáº¿u giá»›i tÃ­nh)
    if len(parts) < 6:
        print(f"âš ï¸ QR chá»‰ cÃ³ {len(parts)} pháº§n, cáº§n Ã­t nháº¥t 6 pháº§n!")
        return None

    # Äáº£m báº£o Ä‘á»§ 7 pháº§n (thÃªm empty náº¿u thiáº¿u)
    while len(parts) < 7:
        parts.append("")

    id_number = parts[0].strip()
    so = parts[1].strip()
    ho_ten = parts[2].strip()
    ngay_sinh = parts[3].strip()
    gioi_tinh = parts[4].strip()

    # Xá»­ lÃ½ trÆ°á»ng há»£p giá»›i tÃ­nh bá»‹ lá»—i
    if gioi_tinh and 'ï¿½' in gioi_tinh:
        # TÃ¡ch Ä‘á»‹a chá»‰ ra khá»i giá»›i tÃ­nh náº¿u bá»‹ dÃ­nh
        parts_gt = gioi_tinh.split(',', 1)
        if len(parts_gt) > 1:
            gioi_tinh = parts_gt[0].strip()
            # Pháº§n sau lÃ  Ä‘á»‹a chá»‰
            dia_chi = parts_gt[1].strip() + ', ' + parts[5].strip() if parts[5] else parts_gt[1].strip()
        else:
            # Chá»‰ láº¥y kÃ½ tá»± Ä‘áº§u náº¿u biáº¿t lÃ  N = Ná»¯
            if gioi_tinh.startswith('N'):
                gioi_tinh = 'Ná»¯'
            dia_chi = parts[5].strip()
    else:
        dia_chi = parts[5].strip()

    ngay_cap = parts[6].strip()

    # Äá»‹nh dáº¡ng láº¡i ngÃ y
    def fmt_date(s):
        try:
            return datetime.strptime(s, "%d%m%Y").strftime("%d/%m/%Y")
        except:
            return s

    return {
        "id": id_number,
        "so": so,
        "ho_ten": ho_ten,
        "ngay_sinh": fmt_date(ngay_sinh),
        "gioi_tinh": gioi_tinh,
        "dia_chi": dia_chi,
        "ngay_cap": fmt_date(ngay_cap),
    }

# =========================================================
# ğŸš€ STEP 4: DÃ² tÃ¬m QR vÃ  trÃ­ch xuáº¥t thÃ´ng tin
# =========================================================
image_path = "/content/ID-Card-VietNam-2/test/images/FILEUPLOAD_b94903b3-953d-4fd2-bf0d-181d746ea6b9_202507281045318210.jpg"
assert os.path.exists(image_path), "âš ï¸ KhÃ´ng tÃ¬m tháº¥y file áº£nh test!"
img = cv2.imread(image_path)

# Dá»± Ä‘oÃ¡n
results = model.predict(source=image_path, conf=0.4, iou=0.3)

# =========================================================
# ğŸš€ STEP 5: Láº·p qua cÃ¡c object Ä‘á»ƒ tÃ¬m QR_CODE
# =========================================================
found_qr = False
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])
        label = r.names[cls_id]

        if label.lower() == "qr_code":
            found_qr = True
            xyxy = box.xyxy[0].tolist()
            x1, y1, x2, y2 = xyxy
            w, h = x2 - x1, y2 - y1

            # Cáº¯t vÃ¹ng QR
            qr_text = extract_qr_from_image(img, (x1, y1, w, h))

            if qr_text:
                print("âœ… QR raw:", qr_text)
                info = parse_vietnam_id_qr(qr_text)
                if info:
                    print("\nğŸ“„ ===== THÃ”NG TIN CCCD =====")
                    print(f"ğŸ†” Sá»‘ CCCD: {info['id']}")
                    print(f"ğŸ“‹ Sá»‘: {info['so']}")
                    print(f"ğŸ‘¤ Há» tÃªn: {info['ho_ten']}")
                    print(f"ğŸ‚ NgÃ y sinh: {info['ngay_sinh']}")
                    print(f"âš§ï¸ Giá»›i tÃ­nh: {info['gioi_tinh']}")
                    print(f"ğŸ  Äá»‹a chá»‰: {info['dia_chi']}")
                    print(f"ğŸ“… NgÃ y cáº¥p: {info['ngay_cap']}")
                    print("=" * 40)
                else:
                    print("âš ï¸ KhÃ´ng parse Ä‘Æ°á»£c thÃ´ng tin CCCD!")
            else:
                print("âš ï¸ KhÃ´ng Ä‘á»c Ä‘Æ°á»£c QR code, thá»­ áº£nh khÃ¡c hoáº·c rÃµ hÆ¡n!")

if not found_qr:
    print("âš ï¸ KhÃ´ng phÃ¡t hiá»‡n Ä‘Æ°á»£c vÃ¹ng QR trÃªn áº£nh!")
```
