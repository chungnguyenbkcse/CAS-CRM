```python
import cv2
import numpy as np
from roboflow import Roboflow
from qreader import QReader
from datetime import datetime
from PIL import Image

# =========================================================
# üöÄ STEP 2: Load model
# =========================================================
model_path = "/content/runs/detect/train/weights/best.pt"
assert os.path.exists(model_path), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y file model best.pt!"
model = YOLO(model_path)

# --- H√†m ƒë·ªçc QR ---
def extract_qr_from_image(img, box):
    x, y, w, h = box
    crop = img[int(y):int(y+h), int(x):int(x+w)]
    qreader = QReader()
    decoded_text = qreader.detect_and_decode(image=crop)
    return decoded_text

# --- H√†m format ng√†y t·ª´ QR ---
def extract_and_format_date_from_qr(qr_text):
    if not qr_text:
        return None
    try:
        date_string = ''.join(c for c in qr_text.split('|')[-1] if c.isdigit())
        date_object = datetime.strptime(date_string, '%d%m%Y')
        return date_object.strftime('%d/%m/%Y')
    except Exception as e:
        print("L·ªói parse ng√†y t·ª´ QR:", e)
        return None

# --- Test ---
image_path = "/content/ID-Card-VietNam-2/test/images/FILEUPLOAD_0a44927d-48f9-40ac-9846-07835c66418e_202501091415548390_jpg.rf.e218542bee56298eef4ee5669e34e74c.jpg"
img = cv2.imread(image_path)
pred = model.predict(image_path, confidence=50).json()

for det in pred["predictions"]:
    if det["class"].lower() == "qr_code":
        x, y, w, h = det["x"] - det["width"]/2, det["y"] - det["height"]/2, det["width"], det["height"]
        qr_text = extract_qr_from_image(img, (x, y, w, h))
        print("QR raw:", qr_text)
        formatted_date = extract_and_format_date_from_qr(qr_text)
        print("Ng√†y c·∫•p:", formatted_date)
```

```python
import cv2
from PIL import Image
from ultralytics import YOLO
import numpy as np

def safe_decode(qr_data_bytes):
    """Th·ª≠ decode d·ªØ li·ªáu QR b·∫±ng nhi·ªÅu b·∫£ng m√£ ƒë·ªÉ tr√°nh l·ªói font."""
    encodings = ['utf-8', 'windows-1258', 'latin1', 'big5', 'shift-jis']
    for enc in encodings:
        try:
            return qr_data_bytes.decode(enc)
        except:
            continue
    return qr_data_bytes.decode('utf-8', errors='ignore')

def detect_and_parse(image_path, model_path):
    # Load model YOLO
    model = YOLO(model_path)

    # ƒê·ªçc ·∫£nh
    img = Image.open(image_path)

    # Detect objects
    results = model.predict(source=image_path, conf=0.5, verbose=False)
    print(results[0].verbose())

    # D√πng OpenCV ƒë·ªçc QR
    img_cv = cv2.imread(image_path)
    detector = cv2.QRCodeDetector()

    data, points, _ = detector.detectAndDecode(img_cv)
    if not data:
        print("‚ö†Ô∏è Kh√¥ng ph√°t hi·ªán m√£ QR n√†o.")
        return

    # N·∫øu l·ªói font, th·ª≠ decode l·∫°i
    try:
        qr_raw = safe_decode(data.encode('latin1'))
    except Exception as e:
        qr_raw = data
        print("‚ö†Ô∏è Kh√¥ng th·ªÉ decode l·∫°i:", e)

    print("‚úÖ QR raw:", qr_raw)

    # Parse theo ƒë·ªãnh d·∫°ng CCCD Vi·ªát Nam (th∆∞·ªùng 7 tr∆∞·ªùng, c√°ch nhau b·∫±ng '|')
    parts = qr_raw.split('|')
    if len(parts) == 7:
        info = {
            'id': parts[0],
            'so': parts[1],
            'ho_ten': parts[2],
            'ngay_sinh': parts[3],
            'gioi_tinh': parts[4],
            'dia_chi': parts[5],
            'ngay_cap': parts[6]
        }
        print("üìÑ Th√¥ng tin CCCD:", info)
    else:
        print("‚ö†Ô∏è QR kh√¥ng ƒë√∫ng ƒë·ªãnh d·∫°ng CCCD Vi·ªát Nam!")
        print("üìÑ Th√¥ng tin CCCD:", None)

# === G·ªåI H√ÄM ===
detect_and_parse(
    image_path="/content/ID-Card-VietNam-2/test/images/cccd_me.jpg",
    model_path="/content/runs/detect/train/weights/best.pt"
)
```

```python
import cv2
import os
from ultralytics import YOLO
from qreader import QReader
from datetime import datetime
from PIL import Image

# =========================================================
# üöÄ STEP 3: Load model YOLO ƒë√£ train
# =========================================================
model_path = "/content/runs/detect/train/weights/best.pt"  # ƒê∆∞·ªùng d·∫´n t·ªõi model ƒë√£ train
assert os.path.exists(model_path), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y file model best.pt!"
model = YOLO(model_path)

# =========================================================
# üß† H√ÄM 1: C·∫Øt v√πng QR t·ª´ ·∫£nh theo to·∫° ƒë·ªô YOLO
# =========================================================
def extract_qr_from_image(img, box):
    x, y, w, h = box
    crop = img[int(y):int(y+h), int(x):int(x+w)]
    qreader = QReader()
    decoded_text = qreader.detect_and_decode(image=crop)
    return decoded_text

# =========================================================
# üß† H√ÄM 2: Parse th√¥ng tin CCCD t·ª´ QR code Vi·ªát Nam
# =========================================================
def parse_vietnam_id_qr(qr_text):
    """
    Chu·∫©n QR CCCD g·ªìm 7 ph·∫ßn ngƒÉn c√°ch b·ªüi d·∫•u |
    V√≠ d·ª•: 075183018264|271524678|Nguy·ªÖn Th·ªã Ng·ªçc Thu·∫≠n|06031983|N·ªØ|T·ªï 1, ·∫§p L·ªô 25, H∆∞ng L·ªôc, Th·ªëng Nh·∫•t, ƒê·ªìng Nai|12082021
    """
    if not qr_text:
        return None

    if isinstance(qr_text, tuple):
        qr_text = qr_text[0]  # b√≥c ra ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n

    parts = qr_text.split("|")
    if len(parts) < 7:
        print("‚ö†Ô∏è QR kh√¥ng ƒë√∫ng ƒë·ªãnh d·∫°ng CCCD Vi·ªát Nam!")
        return None

    id_number = parts[0].strip()
    so = parts[1].strip()
    ho_ten = parts[2].strip()
    ngay_sinh = parts[3].strip()
    gioi_tinh = parts[4].strip()
    dia_chi = parts[5].strip()
    ngay_cap = parts[6].strip()

    # ƒê·ªãnh d·∫°ng l·∫°i ng√†y
    def fmt_date(s):
        try:
            return datetime.strptime(s, "%d%m%Y").strftime("%d/%m/%Y")
        except:
            return s

    return {
        "id": id_number,
        "so": so,
        "ho_ten": ho_ten,
        "ngay_sinh": fmt_date(ngay_sinh),
        "gioi_tinh": gioi_tinh,
        "dia_chi": dia_chi,
        "ngay_cap": fmt_date(ngay_cap),
    }

# =========================================================
# üöÄ STEP 4: D√≤ t√¨m QR v√† tr√≠ch xu·∫•t th√¥ng tin
# =========================================================
image_path = "/content/ID-Card-VietNam-2/test/images/FILEUPLOAD_b94903b3-953d-4fd2-bf0d-181d746ea6b9_202507281045318210.jpg"
assert os.path.exists(image_path), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y file ·∫£nh test!"
img = cv2.imread(image_path)

# D·ª± ƒëo√°n
results = model.predict(source=image_path, conf=0.4, iou=0.3)

# =========================================================
# üöÄ STEP 5: L·∫∑p qua c√°c object ƒë·ªÉ t√¨m QR_CODE
# =========================================================
found_qr = False
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])
        label = r.names[cls_id]

        if label.lower() == "qr_code":
            found_qr = True
            xyxy = box.xyxy[0].tolist()
            x1, y1, x2, y2 = xyxy
            w, h = x2 - x1, y2 - y1

            # C·∫Øt v√πng QR
            qr_text = extract_qr_from_image(img, (x1, y1, w, h))

            if isinstance(qr_text, tuple):
                qr_text = qr_text[0]

            if qr_text:
                print("‚úÖ QR raw:", qr_text)
                info = parse_vietnam_id_qr(qr_text)
                print("üìÑ Th√¥ng tin CCCD:", info)
            else:
                print("‚ö†Ô∏è Kh√¥ng ƒë·ªçc ƒë∆∞·ª£c QR code, th·ª≠ ·∫£nh kh√°c ho·∫∑c r√µ h∆°n!")

if not found_qr:
    print("‚ö†Ô∏è Kh√¥ng ph√°t hi·ªán ƒë∆∞·ª£c v√πng QR tr√™n ·∫£nh!")
```

```python
import cv2
import os
from ultralytics import YOLO
from pyzbar.pyzbar import decode as pyzbar_decode
from datetime import datetime
from PIL import Image
import numpy as np

# =========================================================
# üöÄ STEP 3: Load model YOLO ƒë√£ train
# =========================================================
model_path = "/content/runs/detect/train/weights/best.pt"  # ƒê∆∞·ªùng d·∫´n t·ªõi model ƒë√£ train
assert os.path.exists(model_path), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y file model best.pt!"
model = YOLO(model_path)

# =========================================================
# üß† H√ÄM 1: C·∫Øt v√πng QR t·ª´ ·∫£nh theo to·∫° ƒë·ªô YOLO
# =========================================================
def extract_qr_from_image(img, box):
    x, y, w, h = box
    crop = img[int(y):int(y+h), int(x):int(x+w)]

    # M·ªü r·ªông v√πng crop m·ªôt ch√∫t ƒë·ªÉ tr√°nh m·∫•t vi·ªÅn QR
    padding = 10
    y1 = max(0, int(y) - padding)
    y2 = min(img.shape[0], int(y + h) + padding)
    x1 = max(0, int(x) - padding)
    x2 = min(img.shape[1], int(x + w) + padding)
    crop = img[y1:y2, x1:x2]

    # Upscale ·∫£nh QR l√™n 3x ƒë·ªÉ d·ªÖ ƒë·ªçc h∆°n
    scale_factor = 3
    h_new, w_new = crop.shape[0] * scale_factor, crop.shape[1] * scale_factor
    crop_large = cv2.resize(crop, (w_new, h_new), interpolation=cv2.INTER_CUBIC)

    # Th·ª≠ nhi·ªÅu preprocessing methods
    def try_decode(image):
        # Method 1: D√πng pyzbar tr·ª±c ti·∫øp
        decoded_objects = pyzbar_decode(image)
        if decoded_objects:
            return decoded_objects[0].data

        # Method 2: D√πng OpenCV QRCodeDetector
        detector = cv2.QRCodeDetector()
        data, bbox, _ = detector.detectAndDecode(image)
        if data:
            return data.encode('utf-8')

        return None

    # 1. Th·ª≠ v·ªõi ·∫£nh g·ªëc ƒë√£ upscale
    result = try_decode(crop_large)
    if result:
        return decode_qr_data(result)

    # 2. Th·ª≠ v·ªõi grayscale
    gray = cv2.cvtColor(crop_large, cv2.COLOR_BGR2GRAY)
    result = try_decode(gray)
    if result:
        return decode_qr_data(result)

    # 3. Th·ª≠ v·ªõi adaptive threshold
    thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                     cv2.THRESH_BINARY, 11, 2)
    result = try_decode(thresh)
    if result:
        return decode_qr_data(result)

    # 4. Th·ª≠ v·ªõi Otsu threshold
    _, otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    result = try_decode(otsu)
    if result:
        return decode_qr_data(result)

    # 5. Th·ª≠ l√†m s·∫Øc n√©t ·∫£nh
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    sharpened = cv2.filter2D(gray, -1, kernel)
    result = try_decode(sharpened)
    if result:
        return decode_qr_data(result)

    return None

def decode_qr_data(qr_data):
    """Decode QR data bytes sang text v·ªõi encoding ph√π h·ª£p"""
    if isinstance(qr_data, str):
        return qr_data

    # Debug: In ra bytes raw ƒë·ªÉ xem
    print(f"üîç DEBUG - Raw bytes (first 120): {qr_data[:120]}")
    print(f"üîç DEBUG - Hex: {qr_data[:120].hex()}")

    # Chi·∫øn l∆∞·ª£c: Decode UTF-8 v·ªõi replace ƒë·ªÉ gi·ªØ c·∫•u tr√∫c
    # Thay k√Ω t·ª± l·ªói b·∫±ng ÔøΩ r·ªìi x·ª≠ l√Ω sau
    decoded = qr_data.decode('utf-8', errors='replace')
    print(f"üîç UTF-8 v·ªõi replace: {decoded}")

    # N·∫øu c√≥ k√Ω t·ª± replacement ÔøΩ, th·ª≠ fix b·∫±ng c√°ch decode l·∫°i nh·ªØng ph·∫ßn c√≥ th·ªÉ
    if 'ÔøΩ' in decoded:
        # T√°ch theo d·∫•u | tr∆∞·ªõc
        parts = decoded.split('|')
        fixed_parts = []

        for i, part in enumerate(parts):
            if 'ÔøΩ' in part:
                # Th·ª≠ decode ph·∫ßn n√†y v·ªõi c√°c encoding kh√°c
                # T√¨m v·ªã tr√≠ trong bytes g·ªëc
                start_idx = qr_data.find(part.encode('utf-8', errors='ignore'))

                # V·ªõi ph·∫ßn gi·ªõi t√≠nh th∆∞·ªùng l√† "Nam" ho·∫∑c "N·ªØ"
                if i == 4 and part.startswith('N'):
                    # Th·ª≠ thay th·∫ø
                    if part.strip().startswith('N'):
                        fixed_parts.append('N·ªØ')  # Gi·∫£ ƒë·ªãnh l√† N·ªØ
                        print(f"üîß Fixed part {i}: 'N?' -> 'N·ªØ'")
                        continue

                # Gi·ªØ nguy√™n ph·∫ßn c√≥ ÔøΩ, s·∫Ω fix sau
                fixed_parts.append(part)
            else:
                fixed_parts.append(part)

        result = '|'.join(fixed_parts)
        print(f"‚úÖ Decoded result: {result}")
        return result

    return decoded

# =========================================================
# üß† H√ÄM 2: Parse th√¥ng tin CCCD t·ª´ QR code Vi·ªát Nam
# =========================================================
def parse_vietnam_id_qr(qr_text):
    """
    Chu·∫©n QR CCCD g·ªìm 7 ph·∫ßn ngƒÉn c√°ch b·ªüi d·∫•u |
    V√≠ d·ª•: 075183018264|271524678|Nguy·ªÖn Th·ªã Ng·ªçc Thu·∫≠n|06031983|N·ªØ|T·ªï 1, ·∫§p L·ªô 25, H∆∞ng L·ªôc, Th·ªëng Nh·∫•t, ƒê·ªìng Nai|12082021
    """
    if not qr_text:
        return None

    if isinstance(qr_text, tuple):
        qr_text = qr_text[0]  # b√≥c ra ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n

    parts = qr_text.split("|")
    print(f"üîç S·ªë ph·∫ßn t√°ch ƒë∆∞·ª£c: {len(parts)}")
    print(f"üîç C√°c ph·∫ßn: {parts}")

    # Linh ho·∫°t: ch·∫•p nh·∫≠n 6-7 ph·∫ßn (c√≥ th·ªÉ thi·∫øu gi·ªõi t√≠nh)
    if len(parts) < 6:
        print(f"‚ö†Ô∏è QR ch·ªâ c√≥ {len(parts)} ph·∫ßn, c·∫ßn √≠t nh·∫•t 6 ph·∫ßn!")
        return None

    # ƒê·∫£m b·∫£o ƒë·ªß 7 ph·∫ßn (th√™m empty n·∫øu thi·∫øu)
    while len(parts) < 7:
        parts.append("")

    id_number = parts[0].strip()
    so = parts[1].strip()
    ho_ten = parts[2].strip()
    ngay_sinh = parts[3].strip()
    gioi_tinh = parts[4].strip()

    # X·ª≠ l√Ω tr∆∞·ªùng h·ª£p gi·ªõi t√≠nh b·ªã l·ªói
    if gioi_tinh and 'ÔøΩ' in gioi_tinh:
        # T√°ch ƒë·ªãa ch·ªâ ra kh·ªèi gi·ªõi t√≠nh n·∫øu b·ªã d√≠nh
        parts_gt = gioi_tinh.split(',', 1)
        if len(parts_gt) > 1:
            gioi_tinh = parts_gt[0].strip()
            # Ph·∫ßn sau l√† ƒë·ªãa ch·ªâ
            dia_chi = parts_gt[1].strip() + ', ' + parts[5].strip() if parts[5] else parts_gt[1].strip()
        else:
            # Ch·ªâ l·∫•y k√Ω t·ª± ƒë·∫ßu n·∫øu bi·∫øt l√† N = N·ªØ
            if gioi_tinh.startswith('N'):
                gioi_tinh = 'N·ªØ'
            dia_chi = parts[5].strip()
    else:
        dia_chi = parts[5].strip()

    ngay_cap = parts[6].strip()

    # ƒê·ªãnh d·∫°ng l·∫°i ng√†y
    def fmt_date(s):
        try:
            return datetime.strptime(s, "%d%m%Y").strftime("%d/%m/%Y")
        except:
            return s

    return {
        "id": id_number,
        "so": so,
        "ho_ten": ho_ten,
        "ngay_sinh": fmt_date(ngay_sinh),
        "gioi_tinh": gioi_tinh,
        "dia_chi": dia_chi,
        "ngay_cap": fmt_date(ngay_cap),
    }

# =========================================================
# üöÄ STEP 4: D√≤ t√¨m QR v√† tr√≠ch xu·∫•t th√¥ng tin
# =========================================================
image_path = "/content/ID-Card-VietNam-2/test/images/FILEUPLOAD_b94903b3-953d-4fd2-bf0d-181d746ea6b9_202507281045318210.jpg"
assert os.path.exists(image_path), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y file ·∫£nh test!"
img = cv2.imread(image_path)

# D·ª± ƒëo√°n
results = model.predict(source=image_path, conf=0.4, iou=0.3)

# =========================================================
# üöÄ STEP 5: L·∫∑p qua c√°c object ƒë·ªÉ t√¨m QR_CODE
# =========================================================
found_qr = False
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])
        label = r.names[cls_id]

        if label.lower() == "qr_code":
            found_qr = True
            xyxy = box.xyxy[0].tolist()
            x1, y1, x2, y2 = xyxy
            w, h = x2 - x1, y2 - y1

            # C·∫Øt v√πng QR
            qr_text = extract_qr_from_image(img, (x1, y1, w, h))

            if qr_text:
                print("‚úÖ QR raw:", qr_text)
                info = parse_vietnam_id_qr(qr_text)
                if info:
                    print("\nüìÑ ===== TH√îNG TIN CCCD =====")
                    print(f"üÜî S·ªë CCCD: {info['id']}")
                    print(f"üìã S·ªë: {info['so']}")
                    print(f"üë§ H·ªç t√™n: {info['ho_ten']}")
                    print(f"üéÇ Ng√†y sinh: {info['ngay_sinh']}")
                    print(f"‚ößÔ∏è Gi·ªõi t√≠nh: {info['gioi_tinh']}")
                    print(f"üè† ƒê·ªãa ch·ªâ: {info['dia_chi']}")
                    print(f"üìÖ Ng√†y c·∫•p: {info['ngay_cap']}")
                    print("=" * 40)
                else:
                    print("‚ö†Ô∏è Kh√¥ng parse ƒë∆∞·ª£c th√¥ng tin CCCD!")
            else:
                print("‚ö†Ô∏è Kh√¥ng ƒë·ªçc ƒë∆∞·ª£c QR code, th·ª≠ ·∫£nh kh√°c ho·∫∑c r√µ h∆°n!")

if not found_qr:
    print("‚ö†Ô∏è Kh√¥ng ph√°t hi·ªán ƒë∆∞·ª£c v√πng QR tr√™n ·∫£nh!")
```
